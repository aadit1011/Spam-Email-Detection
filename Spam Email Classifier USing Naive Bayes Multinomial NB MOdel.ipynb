```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv('spam.csv')
df 
```

```python
df.groupby('Category').describe()
```

```python
df['Category']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
df 
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_text,y_train,y_test=train_test_split(df['Message'],df['Category'],train_size=0.75)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
x_train_cv=cv.fit_transform(x_train)
x_train_cv
```

```python
x_train_cv.toarray()[:,3]
```

```python
cv.get_feature_names_out()
```

```python
print(x_train[:10])  # look at first 10 samples
```

```python
from sklearn.naive_bayes import MultinomialNB

mn=MultinomialNB()
mn.fit(x_train_cv,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Accuracy Score=',accuracy_score(y_test,(mn.predict(cv.transform(x_text)))))
```
